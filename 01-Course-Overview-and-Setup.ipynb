# Course Overview and Setup
## Machine Learning in Production
### MLflow and Model Deployment

In this course data scientists and data engineers learn the best practices for putting machine learning models into production.  It starts with managing experiments, projects, and models using MLflow.  It then explores various deployment options including batch, continuous with Spark Streaming, and on demand with RESTful and containerized services.  Finally, it covers monitoring machine learning models once they have been deployed into production.

By the end of this course, you will have built the infrastructure to log, deploy, and monitor machine learning models.

** Prerequisites:**<br><br>

- Python (`pandas`, `sklearn`, `numpy`)
- Background in machine learning and data science
- Basic knowledge of Spark DataFrames (recommended)

** The course consists of the following lessons:**<br><br>

0. Course Overview and Setup
0. Experiment Tracking
0. Packaging ML Projects
0. Model Management 
0. Production Issues
0. Batch Deployment
0. Streaming Deployment
0. Real Time Deployment
0. Drift Monitoring
0. Alerting
0. Delta Time Travel
